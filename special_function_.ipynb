```python
def add(x,y):
    return x+y
```

```python
add(10,20)
```

```python
fun=lambda a,b : a+b
```

```python
fun(10,30)
```

```python
fun(5,10)
```

```python
oe=lambda x: 'even' if x%2==0 else 'odd' 
```

```python
oe(10)
```

```python
oe(11)
```

```python
funny=lambda m : m**2 if m%2==0 else m**3
```

```python
funny(10)
```

```python
funny(5)
```

```python
# range ---0 to 5000 
```

```python
list(filter(lambda x: x%50==0,range(5000)))
```

```python
l=[10,20,30,50,60]
l1=[1,2,3,4,5]
l2=[100,200,300,400,500]
```

```python
for i,k,j in zip(l,l1,l2):
    print(i*k*j)
```
